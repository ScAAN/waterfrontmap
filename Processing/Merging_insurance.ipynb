```python
# import stuff
import os
os.chdir("C:\\Users\\Maija\\Documents\\Python\\waterfrontmap")
cwd = os.getcwd()
print(cwd)
import pandas as pd 
import numpy as np
import json
from pandas.api.types import is_string_dtype
from pandas.api.types import is_numeric_dtype
```

```python
file_name = 'NotYetConverted\\UNINSURED_CD.csv'
df = pd.read_csv(file_name)
df
```

```python
file_name = 'NotYetConverted\\community_districts.geojson'
#json = pd.read_json(file_name)
data = json.load(open(file_name))
```

```python
data['features'][10]['properties']
```

```python
njson = len(data['features'])

for district in range(njson):
    data['features'][district]['properties']["perc_uninsured"] = float(-10)
    data['features'][district]['properties']["unreliable"] = float(0)
    
for ind, rowCD in enumerate(df["CD"].values):
    if ind>0:
        percent = float(df["Percent (Age"][ind])
        unreliable_flag = float(df["unreliable_flag"][ind])
        for district in range(njson):
            jsonCD = data['features'][district]['properties']["BoroCD"]
            if (jsonCD==rowCD):
                data['features'][district]['properties']["perc_uninsured"] = percent
                data['features'][district]['properties']["unreliable"] = unreliable_flag
            
```

```python
output_file = "uninsured_converted.geojson"
with open(output_file, 'w') as f:
    json.dump(data, f)
```
